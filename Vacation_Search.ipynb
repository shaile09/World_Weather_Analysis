```python
# Get the minmum tempreture 
# Ask the customer to add a minimum and maximum temperature value.Filter the DataFrame for minimum and maximum temperature preferences, and if the rain or snow accumulation is 0 inches or not using conditional statements.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
rain_q = input("Do you want it to be raining? (yes/no)")
snow_q = input("Do you want it to be snowing? (yes/no)")
```

```python
# Filter the dataset to find the cities that fit the criteria.
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp)]
preferred_cities_df.head(10)
```

```python
# conditional statments to filter out raining or snowing 
rainsnow_prefrence = ""

if (rain_q == "no" and snow_q == "no"):
    rainsnow_prefrence = cit_data_df.loc[(city_data_df["Rain inches(last 3hrs)"] <= 0) & \
                                (city_data_df["Snow inches(last 3hrs)"] <= 0)]
    
elif (rain_q == "no" and snow_q == "yes"):
    rainsnow_prefrence = city_data_df.loc[(city_data_df["Rain inches(last 3hrs)"] <= 0) & \
                               (city_data_df["Snow inches(last 3hrs)"] >= 0)]
    
elif (rain_q == "yes" and snow_q == "no"):
    rainsnow_prefrence = city_data_df.loc[(city_data_df["Rain inches(last 3hrs)"] >= 0) & \
                                (city_data_df["Snow inches(last 3hrs)"] <= 0)]
        
else:
    rainsnow_prefrence = city_data_df.loc[(city_data_df["Rain inches(last 3hrs)"] >= 0) & \
                                (city_data_df["Snow inches(last 3hrs)"] >= 0)]

rainsnow_prefrence.head()
```

```python
comp = preferred_cities_df.append(rainsnow_prefrence)
comp
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
Hotelvacation_df = preferred_cities_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()
Hotelvacation_df["Hotel Name"] = ""
```

```python
Hotelvacation_df
```

```python
# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
    "location": "48.8566, 2.3522"}
# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()
```

```python
 # Iterate through the DataFrame.
for index, row in Hotelvacation_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to the params dictionary as values to the location key.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        Hotelvacation_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
# Add a heatmap of temperature for the vacation spots.
locations = Hotelvacation_df[["Lat", "Lng"]]
max_temp = Hotelvacation_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, dissipating=False, max_intensity=300, point_radius=4)

fig.add_layer(heat_layer)
# Call the figure to plot the data.
fig

```
